```python
df.dropna(subset=['nickname'], inplace=True)
```

```python
df['height_cm'].fillna(df['height_cm'].median())
df['height_cm'].isnull().sum()
```

```python
df['weight_in_kg'].fillna(df['weight_in_kg'].median())
df['reach_in_cm'].fillna(df['reach_in_cm'].median())
df['stance'].fillna("Unknown")
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['height_cm']=pd.to_numeric(df['height_cm'])
```

```python
df['weight_in_kg']=pd.to_numeric(df['weight_in_kg'])
```

```python
df
```

```python
Q1 = df['height_cm'].quantile(0.25)
Q3 = df['height_cm'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['height_cm'] >= Q1 - 1.5*IQR) & (df['height_cm'] <= Q3 + 1.5*IQR)]
```

```python
df.shape
```

```python
df.hist()
```

```python
df['height_cm'].plot(kind='hist',bins=20,title='Distribution of Height',figsize=(15,5))
```

```python
ax=df['significant_strike_defence'].plot(kind='hist',title='Significant Strike Defence')
ax.set_xlabel('Strike Defence')

```

```python
cols = ['height_cm', 'weight_in_kg', 'reach_in_cm', 'wins', 'losses','significant_strikes_landed_per_minute', 'significant_striking_accuracy','significant_strikes_absorbed_per_minute','significant_strike_defence']	
for i in cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[i], bins=20, kde=True)
    plt.title(f'Distribution of {i}')
    plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation between numerical features")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df['weight_in_kg'])
plt.title("Boxplot of Fighter weights (kg)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='stance', data=df)
plt.title("Count of fighters by stance")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='height_cm',y='reach_in_cm',data=df)
plt.title("Reach vs Height")
plt.xlabel("Height(cm)")
plt.ylabel("reach(cm)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='weight_in_kg',y='wins',hue='stance', data=df)
plt.title("wins vs weight by stance")
plt.xlabel("weight(kg)")
plt.ylabel("wins")
plt.show()
```

```python
sns.pairplot(df[['height_cm','weight_in_kg','reach_in_cm','wins','losses']])
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.regplot(x='height_cm',y='wins',data=df)
plt.title("height vs Wins")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.regplot(x='reach_in_cm',y='wins',data=df)
plt.title("reach vs Wins")
plt.show()
```

```python
df['stance'].value_counts()
```

```python
top_fighters=df.sort_values(by='wins',ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x='wins',y='name',data=top_fighters)
plt.title("Top 10 fighters by wins")
plt.xlabel("Number of wins")
plt.ylabel("fighter names")
plt.show()
```
